## Data Preparation


Let's identify categorical and continuous data set 

```python
# Check season count

bikeshare_data['season'].value_counts()
```

```python
# Replace season int values with str values (1:spring, 2:summer, 3:fall, 4:winter)

bikeshare_data['season']=bikeshare_data['season'].map({1:'spring', 2:'summer', 3:'fall', 4:'winter'})
```

```python
# Checking the entire data 
bikeshare_data
```

```python
# Replace month value i.e 1:jan,2:feb,3:mar,4:apr,5:may,6:jun,7:jul,8:aug,9:sep,10:oct,11:nov,12:dec

bikeshare_data['mnth']=bikeshare_data['mnth'].map({1:'jan',2:'feb',3:'mar',4:'apr',5:'may',6:'jun',7:'jul',8:'aug',9:'sep',10:'oct',11:'nov',12:'dec'})
```

```python
# Again checking the value 'mnth'
bikeshare_data['mnth'].value_counts()
```

```python
# Replace weekday values(0:'sun',1:'mon',2:'tue',3:'wed',4:'thu',5:'fri',6:'sat')

bikeshare_data['weekday']=bikeshare_data['weekday'].map({0:'sun',1:'mon',2:'tue',3:'wed',4:'thu',5:'fri',6:'sat'})
```

```python
# Again checking the value 'weekday'
bikeshare_data['weekday'].value_counts()
```

```python
# Replace weathersit values (1: 'clear',2: 'mist',3:'light_snow',4:'heavy_rain')

bikeshare_data['weathersit']=bikeshare_data['weathersit'].map({1: 'clear',2: 'mist',3:'light_snow',4:'heavy_rain'})
```

```python
# Again checking the value 'weathersit'
bikeshare_data['weathersit'].value_counts()
```

```python
# Checking if there is unnecessary variable we do not want to use in our analysis
bikeshare_data.head()
```

```python
#Dropping the unnecessary variables: ‘instant’, ‘dteday’, ‘casual’ and ‘registered’
# as 'instant' is a serial no, 'dteday' contains date in serial order,'cnt' is a combination of both 'casual'+'registered'

rem_columns = ['instant','dteday','casual','registered']

bikeshare_data.drop(columns=rem_columns, inplace=True)
bikeshare_data.head()
```

```python
# # Drop the first column for dummy variables created for 'weekday'
wkday=pd.get_dummies(bikeshare_data['weekday'],drop_first=True)

# Add results to the original dataframe

bikeshare_data=pd.concat([bikeshare_data,wkday], axis=1)
bikeshare_data.columns
```

## Splitting the data into Training and Testing sets

```python
from sklearn.model_selection import train_test_split

# In order for the train and test data set to be always of the same rows, we specify this accordingly

df_train,df_test=train_test_split(bikeshare_data, train_size=0.7,test_size=0.3,random_state=100)
```

## Rescaling the features

I will use Min Max scaling

```python
from sklearn.preprocessing import MinMaxScaler
scaler=MinMaxScaler()
```

```python
# Apply scaler() to all the except the dummy variables and Binary Variables

num_vars=['temp','atemp','hum','windspeed','cnt']

df_train[num_vars]=scaler.fit_transform(df_train[num_vars])

df_train.head()
```

```python
# Let's check the correlation coefficients to see which variables are highly correlated

plt.figure(figsize = (25, 15))
sns.heatmap(df_train.corr(), annot = True, cmap='RdYlGn')
plt.show()
```

### In the above Heat Map, the variable 'temp' and 'atemp' are highly positively correlated 

## Dividing into X and Y sets for Data Modelling

```python
y_train=df_train.pop('cnt')
X_train=df_train
```

## Building Model

Now I will use LinearRegression function from Scikit Learn for its compatibility with RFE

```python
# Import RFE and LinearRegression

from sklearn.feature_selection import RFE
from sklearn.linear_model import LinearRegression
```

```python
# Running RFE with output number of variable equal to 13

lm=LinearRegression()
lm.fit(X_train,y_train)

# Running RFE

rfe=RFE(lm,16)

rfe=rfe.fit(X_train,y_train)
```

```python
list(zip(X_train.columns,rfe.support_,rfe.ranking_))
```

```python
# We need only those columns which are RFE supported
col = X_train.columns[rfe.support_]
col
```

## Building model using statsmodel for the detailed statistics

```python
# Create X_train_rfe dataframe with RFE selected variables

X_train_rfe=X_train[col]
```

```python
# Adding a constant variable

import statsmodels.api as sm

X_train_rfe=sm.add_constant(X_train_rfe)
```

```python
# Running the linear model
lm=sm.OLS(y_train,X_train_rfe).fit()
```

```python
# Summary of linear model

print(lm.summary())
```

### Since 'feb' is insignificant in presence of other variable as it's p-value is high, so I have to drop it

```python
X_train_new=X_train_rfe.drop(['feb'],axis=1)
```

### Rebuilding the model without 'feb'

```python
# Add a constant variable
import statsmodels.api as sm

X_train_lm=sm.add_constant(X_train_new)
```

```python
# Running the model

lm=sm.OLS(y_train,X_train_lm).fit()
```

```python
# Summary of linear model

print(lm.summary())
```

```python
# Checking the columns
X_train_new.columns
```

```python
# Dropping 'const'
X_train_m1=X_train_new.drop(['const'],axis =1)
```

```python
# Calculate the VIF of the new model

from statsmodels.stats.outliers_influence import variance_inflation_factor

vif=pd.DataFrame()
X=X_train_m1
vif['Features']=X.columns
vif['VIF']=[variance_inflation_factor(X.values,i) for i in range (X.shape[1])]
vif['VIF']=round(vif['VIF'],2)
vif=vif.sort_values(by='VIF',ascending=False)
vif

```

## Residual Analysis of the train data

So, now to check if the error terms are also normally distributed (which is infact, one of the major assumptions of linear regression), let us plot the histogram of the error terms and see what it looks like.

```python
y_train_cnt = lm.predict(X_train_lm)
```

```python
# Importing the required libraries for plots.
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
# Plot the histogram of the error terms
fig = plt.figure()
sns.distplot((y_train - y_train_cnt), bins = 20)
fig.suptitle('Error Terms', fontsize = 20)                  # Plot heading 
plt.xlabel('Errors', fontsize = 18)                         # X-label
```

## Making Predictions
Applying the scaling on the test sets

```python
num_vars = ['temp', 'atemp', 'hum', 'windspeed', 'cnt']
df_test[num_vars] = scaler.transform(df_test[num_vars])
```

```python
#Dividing into X_test and y_test
y_test = df_test.pop('cnt')
X_test = df_test
```

```python
# Now let's use the model to make predictions

# Creating X_test_new dataframe by dropping variables from X_test
X_test_new = X_test[X_train_new.columns]

# Adding a constant variable 
X_test_new = sm.add_constant(X_test_new)
```

```python
# Making predictions
y_pred = lm.predict(X_test_new)
```

## Calculating the R-squared score on the test set.
where y_test is the test data set for the target variable, and y_pred is the variable containing the predicted values of the target variable on the test set.

```python
from sklearn.metrics import r2_score
r2_score(y_test, y_pred)
```
